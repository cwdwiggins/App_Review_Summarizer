### Import the data

```python
# Data handling and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import data and change data types
reviews = pd.read_csv("reviews_prep.csv")
reviews = reviews.drop(['Unnamed: 0'], axis = 1)
reviews['date'] = pd.to_datetime(reviews['date'])
reviews['review'] = reviews['review'].astype('string')
reviews['title'] = reviews['title'].astype('string')
reviews['review_tokens'] = reviews['review_tokens'].astype("string")
reviews['review_POSes'] = reviews['review_POSes'].astype('string')
reviews['title_tokens'] = reviews['title_tokens'].astype('string')
reviews['title_POSes'] = reviews['title_POSes'].astype('string')
reviews.head(5)
```

Union will only return sorted unique values, it won't take into account value counts

```python
# Create series of reviews segmented by rating
all_reviews_2 = reviews[reviews['rating'] == 2]['review_tokens']
all_reviews_3 = reviews[reviews['rating'] == 3]['review_tokens']
all_reviews_4 = reviews[reviews['rating'] == 4]['review_tokens']
all_reviews_5 = reviews[reviews['rating'] == 5]['review_tokens']

# Create texts
all_text_2 = ' '.join(word for word in all_reviews_2)
all_text_3 = ' '.join(word for word in all_reviews_3)
all_text_4 = ' '.join(word for word in all_reviews_4)
all_text_5 = ' '.join(word for word in all_reviews_5)

# Create DataFrames for each text
all_text_1_df = pd.DataFrame({'word': all_text_1.split(' '), 'count': [1]*len(all_text_1.split(' '))})
all_text_2_df = pd.DataFrame({'word': all_text_2.split(' '), 'count': [1]*len(all_text_2.split(' '))})
all_text_3_df = pd.DataFrame({'word': all_text_3.split(' '), 'count': [1]*len(all_text_3.split(' '))})
all_text_4_df = pd.DataFrame({'word': all_text_4.split(' '), 'count': [1]*len(all_text_4.split(' '))})
all_text_5_df = pd.DataFrame({'word': all_text_5.split(' '), 'count': [1]*len(all_text_5.split(' '))})

# Create series for each text
text_1_Series = all_text_1_df['word']
text_2_Series = all_text_2_df['word']
text_3_Series = all_text_3_df['word']
text_4_Series = all_text_4_df['word']
text_5_Series = all_text_5_df['word']
```

```python
# Find all items that are not in common between all series
from functools import reduce

all_text_union = reduce(np.union1d, (text_1_Series, text_2_Series, text_3_Series, text_4_Series, text_5_Series))
all_text_intersect = reduce(np.intersect1d, (text_1_Series, text_2_Series, text_3_Series, text_4_Series, text_5_Series))

boolmask = np.isin(all_text_union, all_text_intersect)
notcommonarray = all_text_union[~boolmask]
```

```python
# Aggregate the word counts together
word_counts_1 = all_text_1_df.groupby('word')['count'].agg('sum')
word_counts_2 = all_text_2_df.groupby('word')['count'].agg('sum')
word_counts_3 = all_text_3_df.groupby('word')['count'].agg('sum')
word_counts_4 = all_text_4_df.groupby('word')['count'].agg('sum')
word_counts_5 = all_text_5_df.groupby('word')['count'].agg('sum')

# Now use this to filter for words that are just in each rating category
word_counts_1_unique = word_counts_1[word_counts_1.index.isin(notcommonarray)]
word_counts_2_unique = word_counts_2[word_counts_2.index.isin(notcommonarray)]
word_counts_3_unique = word_counts_3[word_counts_3.index.isin(notcommonarray)]
word_counts_4_unique = word_counts_4[word_counts_4.index.isin(notcommonarray)]
word_counts_5_unique = word_counts_5[word_counts_5.index.isin(notcommonarray)]
```

```python
# Arragne and select the top 20 words and their frequencies for each category

def top20wordsXandY(series):
    series_sorted = series.sort_values(ascending = False)
    top20 = series_sorted[0:20]
    top20_ascending = top20.sort_values(ascending = True)
    index = top20_ascending.index
    values = top20_ascending.values
    return index, values
 
# Visualize the frequency plots

fig, ax = plt.subplots(3, 2, figsize = (10, 10))
plt.subplot(3,2,1)
x1, y1 = top20wordsXandY(word_counts_1_unique)
plt.barh(x1, y1, color = 'blue')

plt.subplot(3,2,2)
x2, y2 = top20wordsXandY(word_counts_2_unique)
plt.barh(x2, y2, color = 'red')

plt.subplot(3,2,3)
x3, y3 = top20wordsXandY(word_counts_3_unique)
plt.barh(x3, y3, color = 'orange')

plt.subplot(3,2,4)
x4, y4 = top20wordsXandY(word_counts_4_unique)
plt.barh(x4, y4, color = 'green')

plt.subplot(3,2,5)
x5, y5 = top20wordsXandY(word_counts_5_unique)
plt.barh(x5, y5, color = 'yellow')

fig.supxlabel('frequency')
fig.supylabel('word')
fig.suptitle('Word frequency count for each category')
plt.tight_layout()
plt.show()
```

Now that only the unique words for each category have been kept, we are starting to see some signal here.
